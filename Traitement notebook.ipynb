## Distances marqueurs bas

```python
df_data = pd.read_csv('1-Distance_2marqueurs_bas.csv')

y_data = df_data.Value
y_data_liste = list(y_data)

x_data = df_data.Pixels
x_data_liste = list(x_data)

# normalisons
range_x = max(x_data_liste) - min(x_data_liste)

new_x = []
for i in range(len(x_data_liste)):
    new_x.append(x_data_liste[i]/range_x)

# figure avec l'axe des x (pixels) normalisÃ©
#fig = px.line(df_data, x = new_x, y = 'Value', title='IntensitÃ© en niveaux de gris du trait (en pixels) reliant les deux marqueurs du bas du fantÃ´me. 2D', labels={"x":"Pixels", "Value":"IntensitÃ© en niveau de gris [cd]"})
#fig.show()
plt.plot(new_x, y_data_liste)
plt.title("IntensitÃ© en niveaux de gris du trait (en pixels) reliant les deux marqueurs du bas du fantÃ´me. Image 2D")
plt.xlabel("Pixels")
plt.ylabel("IntensitÃ© en niveau de gris [cd]")
plt.grid()
plt.show()

# je trouve les extremums de chaque pic
y_data_500_1 = []
for i in range(500):
    y_data_500_1.append(y_data_liste[i])
y_min_1 = min(y_data_500_1)
print("le premier minimum est", y_min_1)

y_data_500_2 = []
for i in range(2700,3200):
    y_data_500_2.append(y_data_liste[i])
y_min_2 = min(y_data_500_2)
print("le deuxiÃ¨me minimum est", y_min_2)

# j'extrais la position en x de chaque pic
pos_x_min_1 = y_data_liste.index(y_min_1)
pos_x_min_2 = y_data_liste.index(y_min_2)

# j'extrais la valeur en x de chaque pic
x_value_at_min_1 = x_data_liste[pos_x_min_1]/range_x
x_value_at_min_2 = x_data_liste[pos_x_min_2]/range_x

# je calcule la distance entre les deux pics
dist_reference = x_value_at_min_2 - x_value_at_min_1
print("la distance entre les deux minimums est", dist_reference, "pixels")

# je calcule les nouvelles incertitudes reliÃ©es aux nouvelles valeurs et aux mesures de distance
incertitude_ref = 0.05/max(x_data_liste)
incertitude_distance_ref = np.sqrt((incertitude_ref**2) + (incertitude_ref**2))
print("la nouvelle incertitude sur chaque mesure est", incertitude_ref)
print("la nouvelle incertitude sur la distance est", incertitude_distance_ref)
```

## Position des marqueurs avec l'effet du mouvement

### Haut

```python
plt.rcParams["figure.figsize"] = [10, 5]
plt.rcParams["figure.autolayout"] = True

fig, ax1 = plt.subplots()
# graphique respiration haut
df_data = pd.read_csv('9-320Proj_0.5mm_distance_marqueurs_haut_respiration.csv')

y_data = df_data.Value
y_data_liste = list(y_data)
x_data = df_data.Pixels
x_data_liste = list(x_data)

# normalisons
range_x = max(x_data_liste) - min(x_data_liste)
range_y = max(y_data_liste) 

new_y = []
for i in range(len(y_data_liste)):
    new_y.append(y_data_liste[i]/range_y)

new_x = []
for i in range(len(x_data_liste)):
    new_x.append(x_data_liste[i]/range_x)

ax1.plot(new_x, new_y, color = "red", label = "Respiration")
plt.legend()
plt.xlabel("Pixels")
plt.ylabel("IntensitÃ© en niveaux de gris [cd]")
#############################

# graphique 2D haut
df_data = pd.read_csv('2-Distance_2marquers_haut.csv')

y_data = df_data.Value
y_data_liste = list(y_data)
x_data = df_data.Pixels
x_data_liste = list(x_data)

# normalisons
range_x = max(x_data_liste) - min(x_data_liste)
range_y = max(y_data_liste) 

new_y = []
for i in range(len(y_data_liste)):
    new_y.append(y_data_liste[i]/range_y)

new_x = []
for i in range(len(x_data_liste)):
    new_x.append(x_data_liste[i]/range_x)

ax2 = ax1.twiny()
ax2.plot(new_x, new_y, color = "blue", label = "2D")
plt.legend(loc = "upper left")
plt.title("Position et intensitÃ© lumineuse des extremums correspondant aux marqueurs du haut du fantÃ´me")
plt.show
```
