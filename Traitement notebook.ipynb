## Distances marqueurs bas

```python
df_data = pd.read_csv('1-Distance_2marqueurs_bas.csv')

fig = px.line(df_data, x = 'Pixels', y = 'Value', title='Distance 2D')
fig.show()

y_data = df_data.Value
y_data_liste = list(y_data)
x_data = df_data.Pixels
x_data_liste = list(x_data)

# normalisons
range_x = max(x_data_liste) - min(x_data_liste)

new_x = []
for i in range(len(x_data_liste)):
    new_x.append(x_data_liste[i]/range_x)

fig = px.line(df_data, x = new_x, y = 'Value', title='Distance 2D')
fig.show()

y_data_500_1 = []
for i in range(500):
    y_data_500_1.append(y_data_liste[i])
y_min_1 = min(y_data_500_1)
print("le premier minimum est", y_min_1)

y_data_500_2 = []
for i in range(2700,3200):
    y_data_500_2.append(y_data_liste[i])
y_min_2 = min(y_data_500_2)
print("le deuxiÃ¨me minimum est", y_min_2)

pos_x_min_1 = y_data_liste.index(y_min_1)
pos_x_min_2 = y_data_liste.index(y_min_2)

x_value_at_min_1 = x_data_liste[pos_x_min_1]/range_x
x_value_at_min_2 = x_data_liste[pos_x_min_2]/range_x

dist_reference = x_value_at_min_2 - x_value_at_min_1
print("la distance entre les deux minimums est", dist_reference, "pixels")
```
