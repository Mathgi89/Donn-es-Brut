## Distances marqueurs bas

```python
df_data = pd.read_csv('1-Distance_2marqueurs_bas.csv')

y_data = df_data.Value
y_data_liste = list(y_data)

x_data = df_data.Pixels
x_data_liste = list(x_data)

# normalisons
range_x = max(x_data_liste) - min(x_data_liste)

new_x = []
for i in range(len(x_data_liste)):
    new_x.append(x_data_liste[i]/range_x)

# figure avec l'axe des x (pixels) normalisé
fig = px.line(df_data, x = new_x, y = 'Value', title='Intensité en niveaux de gris du trait (en pixels) reliant les deux marqueurs du bas du fantôme. 2D', labels={"x":"Pixels", "Value":"Intensité en niveau de gris [cd]"})
fig.show()

# je trouve les extremums de chaque pic
y_data_500_1 = []
for i in range(500):
    y_data_500_1.append(y_data_liste[i])
y_min_1 = min(y_data_500_1)
print("le premier minimum est", y_min_1)

y_data_500_2 = []
for i in range(2700,3200):
    y_data_500_2.append(y_data_liste[i])
y_min_2 = min(y_data_500_2)
print("le deuxième minimum est", y_min_2)

# j'extrais la position en x de chaque pic
pos_x_min_1 = y_data_liste.index(y_min_1)
pos_x_min_2 = y_data_liste.index(y_min_2)

# j'extrais la valeur en x de chaque pic
x_value_at_min_1 = x_data_liste[pos_x_min_1]/range_x
x_value_at_min_2 = x_data_liste[pos_x_min_2]/range_x

# je calcule la distance entre les deux pics
dist_reference = x_value_at_min_2 - x_value_at_min_1
print("la distance entre les deux minimums est", dist_reference, "pixels")

# je calcule les nouvelles incertitudes reliées aux nouvelles valeurs et aux mesures de distance
incertitude_ref = 0.05/max(x_data_liste)
incertitude_distance_ref = np.sqrt((incertitude_ref**2) + (incertitude_ref**2))
print(max(x_data_liste))
print("la nouvelle incertitude sur chaque mesure est", incertitude_ref)
print("la nouvelle incertitude sur la distance est", incertitude_distance_ref)
```
